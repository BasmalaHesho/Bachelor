```python
import os
import subprocess
import cv2
import numpy as np
import torchaudio
from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm
from scipy.io import wavfile
from pyannote.audio import Pipeline
from pydub import AudioSegment
import whisper
import torch


# Initialize models
whisper_model = whisper.load_model("large")

def ensure_ffmpeg_installed():
    try:
        subprocess.run(['ffmpeg', '-version'], capture_output=True, check=True)
    except (subprocess.CalledProcessError, FileNotFoundError):
        print("Installing FFmpeg...")
        subprocess.run(['apt-get', 'update'], check=True)
        subprocess.run(['apt-get', 'install', '-y', 'ffmpeg'], check=True)

def download_video(url, output_file):
    try:
        print("Downloading video...")
        subprocess.run(["pip", "install", "-U", "yt-dlp"], check=True)
        subprocess.run([
            "yt-dlp",
            "--format", "bestvideo[ext=mp4]+bestaudio[ext=m4a]/mp4",
            "--output", output_file,
            "--merge-output-format", "mp4",
            url
        ], check=True, capture_output=True, text=True)

        if not os.path.exists(output_file):
            raise FileNotFoundError(f"Failed to download video to {output_file}")
        print("Video downloaded successfully!")
        return True
    except Exception as e:
        print(f"Error downloading video: {str(e)}")
        return False

def extract_audio(video_path, audio_path):
    try:
        print("Extracting audio...")
        result = subprocess.run([
            "ffmpeg", "-y",
            "-i", video_path,
            "-vn",
            "-acodec", "pcm_s16le",
            "-ar", "16000",
            "-ac", "1",
            "-copyts",
            audio_path
        ], capture_output=True, text=True)

        if not os.path.exists(audio_path) or os.path.getsize(audio_path) == 0:
            print("First attempt failed, trying alternative method...")
            result = subprocess.run([
                "ffmpeg", "-y",
                "-i", video_path,
                "-vn",
                "-acodec", "pcm_s16le",
                "-ar", "16000",
                "-ac", "1",
                "-af", "aresample=16000",
                "-copyts",
                audio_path
            ], capture_output=True, text=True)

        if os.path.exists(audio_path) and os.path.getsize(audio_path) > 0:
            print("Audio extraction successful!")
            return True
        else:
            print(f"Failed to extract audio. FFmpeg output: {result.stderr}")
            return False
    except Exception as e:
        print(f"Error extracting audio: {str(e)}")
        return False

def diarize_and_segment_audio(audio_path, diarization_model="pyannote/speaker-diarization-3.1", output_base="speaker_segments"):
    print("Running diarization...")
    pipeline = Pipeline.from_pretrained(diarization_model, use_auth_token=True)
    diarization = pipeline(audio_path)

    speaker_dirs = {}
    os.makedirs(output_base, exist_ok=True)

    audio = AudioSegment.from_wav(audio_path)

    segments = []

    for turn, _, speaker in diarization.itertracks(yield_label=True):
        start_ms = int(turn.start * 1000)
        end_ms = int(turn.end * 1000)

        speaker_dir = os.path.join(output_base, speaker)
        os.makedirs(speaker_dir, exist_ok=True)
        speaker_dirs[speaker] = speaker_dir

        segment_audio = audio[start_ms:end_ms]
        segment_filename = os.path.join(speaker_dir, f"{start_ms}_{end_ms}.wav")
        segment_audio.export(segment_filename, format="wav")

        segments.append({
            "speaker": speaker,
            "start": turn.start,
            "end": turn.end,
            "audio_file": segment_filename
        })

    print(f"Diarization complete. Found {len(set(s['speaker'] for s in segments))} speakers.")
    return segments

def transcribe_with_whisper(audio_path):
    try:
        result = whisper_model.transcribe(audio_path, language="ar")
        return result['text']
    except Exception as e:
        print(f"Whisper failed on {audio_path}: {e}")
        return ""


def process_segments(video_path, diarized_segments, video_output_dir, frame_output_dir):
    os.makedirs(video_output_dir, exist_ok=True)
    os.makedirs(frame_output_dir, exist_ok=True)

    def process_single_segment(seg):
        start_time = seg['start']
        end_time = seg['end']
        speaker = seg['speaker']

        # Generate unique filenames per speaker segment
        segment_filename = f"{speaker}_{int(start_time*1000)}_{int(end_time*1000)}"

        # Extract video segment
        output_video_path = os.path.join(video_output_dir, f"{segment_filename}.mp4")
        subprocess.run([
            "ffmpeg", "-y",
            "-ss", str(start_time),
            "-t", str(end_time - start_time),
            "-i", video_path,
            "-c:v", "libx264",
            "-preset", "ultrafast",
            "-c:a", "aac",
            output_video_path
        ], check=True, capture_output=True)

        # Extract a single frame from the middle of the segment
        midpoint = (start_time + end_time) / 2
        output_frame_path = os.path.join(frame_output_dir, f"{segment_filename}.jpg")
        subprocess.run([
            "ffmpeg", "-y",
            "-ss", str(midpoint),
            "-i", video_path,
            "-vframes", "1",
            output_frame_path
        ], check=True, capture_output=True)

    with ThreadPoolExecutor() as executor:
        list(tqdm(executor.map(process_single_segment, diarized_segments), total=len(diarized_segments)))

def process_video(url, output_dir):
    try:
        ensure_ffmpeg_installed()

        os.makedirs(output_dir, exist_ok=True)
        print(f"Output directory created: {output_dir}")

        video_output_dir = os.path.join(output_dir, "video_clips")
        audio_output_dir = os.path.join(output_dir, "audio_clips")
        frame_output_dir = os.path.join(output_dir, "video_frames")

        os.makedirs(video_output_dir, exist_ok=True)
        os.makedirs(audio_output_dir, exist_ok=True)
        os.makedirs(frame_output_dir, exist_ok=True)

        print(f"Video output directory: {video_output_dir}")
        print(f"Audio output directory: {audio_output_dir}")
        print(f"Frame output directory: {frame_output_dir}")

        video_path = os.path.join(output_dir, "downloaded_video.mp4")
        if not download_video(url, video_path):
            raise RuntimeError("Failed to download video")

        audio_path = os.path.join(output_dir, "extracted_audio.wav")
        if not extract_audio(video_path, audio_path):
            raise RuntimeError("Failed to extract audio")

        print("Running diarization...")
        diarized_segments = diarize_and_segment_audio(audio_path, output_base=audio_output_dir)

        if not diarized_segments:
            raise RuntimeError("No speech segments detected")

        transcriptions = []

        for seg in diarized_segments:
            whisper_text = transcribe_with_whisper(seg["audio_file"])

            seg["transcription"] = whisper_text
            print(f"Speaker {seg['speaker']} ({seg['start']:.2f}s - {seg['end']:.2f}s): {whisper_text}")
            transcriptions.append(seg)

        # Save speaker-wise transcription
        transcription_file = os.path.join(output_dir, "speaker_transcriptions.txt")
        with open(transcription_file, "w", encoding="utf-8") as f:
            for seg in transcriptions:
                f.write(f"Speaker: {seg['speaker']}\n")
                f.write(f"Time: {seg['start']:.2f} - {seg['end']:.2f} seconds\n")
                f.write(f"Audio File: {seg['audio_file']}\n")
                f.write(f"Transcription: {seg['transcription']}\n\n")

        print(f"Speaker segment transcriptions saved to: {transcription_file}")
        print("Processing and transcription completed successfully!")

        process_segments(
            video_path,
            diarized_segments,
            video_output_dir,
            frame_output_dir
        )

    except Exception as e:
        print(f"Error processing video: {str(e)}")


if __name__ == "__main__":
    video_url = "https://youtu.be/qYtRYFtY7e8?feature=shared"
    output_dir = "/content/drive/MyDrive/Trial41May13"
    process_video(video_url, output_dir)
```
